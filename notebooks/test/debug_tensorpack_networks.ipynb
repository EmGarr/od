# Proof that a kernel in channels first or channels last is the same

```python
import numpy as np
import tensorflow as tf

cl = tf.keras.layers.Conv2D(3, 3, data_format='channels_last')
inputs = np.arange(4 * 4 * 3).reshape(1, 4, 4, 3).astype(np.float32)
cl_out = cl(inputs)
cl_weights = cl.get_weights()

inputs_cf = np.moveaxis(inputs, -1, 1)
cf = tf.keras.layers.Conv2D(3, 3, data_format='channels_first')
out_before_weigths = cf(inputs_cf)
cf.set_weights(cl_weights)
out_after_weights =cf(inputs_cf)
np.testing.assert_array_equal(cf.get_weights()[0], cl.get_weights()[0])
np.testing.assert_array_equal(out_after_weights.numpy(), np.moveaxis(cl_out.numpy(), -1, 1))
```
